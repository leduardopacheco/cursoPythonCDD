## 1. Sintaxe Básica

Python é conhecido por ter uma sintaxe limpa e próxima do inglês. Algumas regras fundamentais:

- **Indentação obrigatória:** blocos de código são delimitados por 4 espaços, não por chaves `{}`.
- **Comentários:** qualquer texto após `#` é ignorado pelo Python — use para explicar o código.
- **Sem ponto-e-vírgula:** cada instrução ocupa uma linha.
- **Sensível a maiúsculas:** `nome` e `Nome` são variáveis diferentes.

````python
# Isso é um comentário — o Python ignora esta linha
# A função print() exibe mensagens na tela

print("Bem-vindo ao curso de Python para Ciência de Dados!")
print("Hoje você dá o primeiro passo para se tornar um cientista de dados.")
````

````python
# Você pode imprimir múltiplas coisas separadas por vírgula
print("Olá", "Mundo")            # separador padrão é espaço
print("Olá", "Mundo", sep="-")   # separador personalizado
print("Linha 1", end=" | ")      # não quebra a linha no final
print("Linha 2")
````

## 2. Variáveis e Tipos de Dados

Uma **variável** é um nome que guarda um valor na memória. Em Python, você não precisa declarar o tipo — ele é inferido automaticamente.

### Tipos primitivos principais

| Tipo | Exemplo | Descrição |
|------|---------|----------|
| `int` | `22` | Número inteiro |
| `float` | `1.75` | Número decimal |
| `str` | `"Carlos"` | Texto (string) |
| `bool` | `True` / `False` | Valor lógico |

````python
# Declarando variáveis — não precisamos informar o tipo explicitamente
nome      = "Carlos"   # str  → texto, sempre entre aspas
idade     = 22         # int  → número inteiro
altura    = 1.75       # float → número com casas decimais
estudante = True       # bool  → True ou False

# type() revela o tipo de cada variável
print("nome     :", nome,      "→ tipo:", type(nome))
print("idade    :", idade,     "→ tipo:", type(idade))
print("altura   :", altura,    "→ tipo:", type(altura))
print("estudante:", estudante, "→ tipo:", type(estudante))
````

````python
# f-strings: a forma mais moderna de montar textos com variáveis
# Coloque f antes das aspas e use {variavel} no texto

print(f"Aluno: {nome}, {idade} anos, altura {altura}m")
print(f"É estudante? {estudante}")

# Você pode fazer cálculos dentro das chaves
preco    = 49.9
desconto = 0.10
print(f"Preço com desconto: R$ {preco * (1 - desconto):.2f}")  # :.2f → 2 casas decimais
````

````python
# Conversão entre tipos (type casting)
# Isso é muito comum ao ler dados de arquivos CSV, que chegam como texto

nota_texto  = "8.5"              # lida de um arquivo: chega como string
nota_numero = float(nota_texto)  # convertemos para float

print(f"Antes : {nota_texto!r}  (tipo: {type(nota_texto).__name__})")
print(f"Depois: {nota_numero}   (tipo: {type(nota_numero).__name__})")

# Outros conversores úteis
print(int(3.9))    # → 3     (trunca, não arredonda)
print(str(100))    # → '100'
print(bool(0))     # → False (zero é False)
print(bool(42))    # → True  (qualquer número != 0 é True)
````

## 3. Operadores

Operadores permitem realizar cálculos e comparações — essenciais para filtrar e transformar dados.

````python
# ── Operadores Aritméticos ─────────────────────────────────────────
a, b = 10, 3

print(f"{a} + {b}  = {a + b}")          # adição
print(f"{a} - {b}  = {a - b}")          # subtração
print(f"{a} * {b}  = {a * b}")          # multiplicação
print(f"{a} / {b}  = {a / b:.4f}")      # divisão real → sempre float
print(f"{a} // {b} = {a // b}")         # divisão inteira (quociente)
print(f"{a} % {b}  = {a % b}")          # módulo (resto da divisão)
print(f"{a} ** {b} = {a ** b}")         # potência
````

````python
# ── Operadores de Comparação ───────────────────────────────────────
# Sempre retornam True ou False

nota = 7.5
print(f"nota >= 7 : {nota >= 7}")    # maior ou igual
print(f"nota == 10: {nota == 10}")   # igualdade (== não é =)
print(f"nota != 10: {nota != 10}")   # diferente
print(f"nota < 5  : {nota < 5}")     # menor que
````

````python
# ── Operadores Lógicos ─────────────────────────────────────────────
# and, or, not — combinam condições

nota       = 7.5
frequencia = 80  # porcentagem

# Para ser aprovado: nota >= 7 E frequência >= 75
aprovado = (nota >= 7) and (frequencia >= 75)
print(f"Aprovado? {aprovado}")

# Recuperação: nota entre 5 e 7 OU frequência abaixo de 75
recuperacao = (5 <= nota < 7) or (frequencia < 75)
print(f"Recuperação? {recuperacao}")

# not inverte o resultado
print(f"NÃO aprovado? {not aprovado}")
````

## 4. Estruturas Condicionais

Com `if`, `elif` e `else`, o programa toma decisões diferentes conforme a situação — como ao classificar dados.

````python
# Estrutura básica: if / else
nota = 7.0

if nota >= 7:
    print("Aprovado")
else:
    print("Reprovado")
````

````python
# Múltiplas condições com elif (else if)
# elif é verificado apenas quando as condições anteriores são falsas

nota = 6.2

if nota >= 9:
    situacao = "Excelente"
elif nota >= 7:
    situacao = "Aprovado"
elif nota >= 5:
    situacao = "Recuperação"
else:
    situacao = "Reprovado"

print(f"Nota {nota} → {situacao}")
````

````python
# Condicional ternária: forma compacta para decisões simples
# Sintaxe: valor_se_verdadeiro if condição else valor_se_falso

nota = 8
resultado = "Aprovado" if nota >= 7 else "Reprovado"
print(resultado)

# Exemplo prático: classificar um valor numérico
temperatura = 38.5
status = "Febre" if temperatura >= 37.5 else "Normal"
print(f"Temperatura {temperatura}°C → {status}")
````

## 5. Laços de Repetição

Laços permitem repetir instruções automaticamente — essencial para processar coleções de dados.

- **`for`** → itera sobre uma sequência conhecida
- **`while`** → repete enquanto uma condição for verdadeira

````python
# For com range(inicio, fim, passo)
# range(1, 6) gera os números 1, 2, 3, 4, 5  (o fim é exclusivo)

print("Contagem:")
for i in range(1, 6):
    print(f"  Número {i}")
````

````python
# Iterando diretamente sobre uma lista
notas = [8.5, 6.0, 9.0, 5.5, 7.5]

soma = 0
for nota in notas:
    soma += nota   # equivale a: soma = soma + nota

media = soma / len(notas)
print(f"Notas: {notas}")
print(f"Soma: {soma} | Média: {media:.2f}")
````

````python
# enumerate() → obtém índice e valor ao mesmo tempo
alunos = ["Ana", "Bruno", "Carla", "Diego", "Elaine"]

for i, aluno in enumerate(alunos, start=1):
    print(f"  {i}. {aluno}")
````

````python
# While: útil quando não sabemos antecipadamente quantas iterações serão necessárias
saldo = 1000.0
mes   = 0
taxa  = 0.02  # 2% ao mês

print("Simulação de crescimento com juros compostos:")
while saldo < 1500:
    saldo *= (1 + taxa)
    mes   += 1
    print(f"  Mês {mes:2d}: R$ {saldo:.2f}")

print(f"\nLevou {mes} meses para passar de R$ 1.500")
````

````python
# break    → interrompe o laço imediatamente
# continue → pula para a próxima iteração

notas = [8.5, 9.0, 4.0, 7.5, 6.0]

print("Verificando a primeira nota abaixo de 5:")
for nota in notas:
    if nota < 5:
        print(f"  → Nota {nota} está abaixo de 5. Parando aqui.")
        break
    print(f"  Nota {nota} OK")

print("\nApenas notas >= 7:")
for nota in notas:
    if nota < 7:
        continue   # pula para a próxima iteração sem imprimir
    print(f"  Nota aprovada: {nota}")
````

````python
# List comprehension: forma compacta e pythônica de criar listas
# Sintaxe: [expressão for item in sequência if condição]

notas = [8.5, 6.0, 9.0, 5.5, 7.5, 4.0]

# Forma tradicional
aprovadas_trad = []
for n in notas:
    if n >= 7:
        aprovadas_trad.append(n)

# Forma compacta (list comprehension) — mesmo resultado em uma linha
aprovadas_comp = [n for n in notas if n >= 7]

print("Tradicional  :", aprovadas_trad)
print("Comprehension:", aprovadas_comp)

# Exemplo: classificar cada nota em uma lista
situacoes = ["Aprovado" if n >= 7 else "Reprovado" for n in notas]
print("Situações    :", situacoes)
````

## 6. Estruturas de Dados

Estruturas de dados organizam informações — são a base antes de trabalharmos com Pandas.

| Estrutura | Símbolo | Ordenada | Mutável | Duplicatas |
|-----------|---------|----------|---------|------------|
| Lista (`list`) | `[]` | ✓ | ✓ | ✓ |
| Tupla (`tuple`) | `()` | ✓ | ✗ | ✓ |
| Dicionário (`dict`) | `{}` | ✓* | ✓ | chaves: ✗ |
| Conjunto (`set`) | `{}` | ✗ | ✓ | ✗ |

*ordenado por inserção desde Python 3.7

````python
# ── LISTA ─────────────────────────────────────────────────────────
# Coleção ordenada e mutável — equivale a uma coluna de dados

notas = [7.5, 8.0, 6.5, 9.0, 5.0]

# Acesso por índice (começa em 0)
print("Primeira nota:", notas[0])    # índice 0
print("Última nota:  ", notas[-1])   # índice negativo = da direita

# Fatia (slicing)
print("Três primeiras:", notas[:3])  # do início até índice 3 (exclusivo)

# Métodos úteis
notas.append(10.0)        # adiciona ao final
notas.sort()              # ordena no lugar
print("Ordenadas:", notas)
print(f"Máx: {max(notas)} | Mín: {min(notas)} | Total: {len(notas)}")
````

````python
# ── DICIONÁRIO ────────────────────────────────────────────────────
# Pares chave → valor — equivale a uma linha de um dataset

aluno = {
    "nome"       : "Marina",
    "idade"      : 19,
    "curso"      : "Ciência de Dados",
    "nota_final" : 8.5,
    "aprovado"   : True
}

# Acessar valores
print("Nome:", aluno["nome"])
print("Nota:", aluno["nota_final"])

# .get() é mais seguro: retorna None (ou um padrão) se a chave não existir
print("Email:", aluno.get("email", "não informado"))

# Adicionar uma nova chave
aluno["frequencia"] = 92
print("\nAluno completo:")
for chave, valor in aluno.items():
    print(f"  {chave}: {valor}")
````

````python
# Lista de dicionários → estrutura próxima de uma tabela (DataFrame)

turma = [
    {"nome": "Ana",    "nota": 8.5, "frequencia": 90},
    {"nome": "Bruno",  "nota": 6.0, "frequencia": 75},
    {"nome": "Carla",  "nota": 9.0, "frequencia": 95},
    {"nome": "Diego",  "nota": 5.5, "frequencia": 68},
    {"nome": "Elaine", "nota": 7.5, "frequencia": 88},
]

print("Alunos aprovados (nota >= 7):")
for aluno in turma:
    if aluno["nota"] >= 7:
        print(f"  ✓ {aluno['nome']} — nota {aluno['nota']}")
````

````python
# ── TUPLA ─────────────────────────────────────────────────────────
# Imutável: boa para dados que não devem mudar (coordenadas, constantes)

coordenada = (-3.7172, -38.5434)  # Fortaleza: (latitude, longitude)
print("Latitude :", coordenada[0])
print("Longitude:", coordenada[1])
# Tentar modificar gera erro — descomente para ver:
# coordenada[0] = 0   # → TypeError!

# ── CONJUNTO ──────────────────────────────────────────────────────
# Sem duplicatas — útil para encontrar valores únicos em dados

categorias_com_repeticao = ["A", "B", "A", "C", "B", "A", "D"]
categorias_unicas = set(categorias_com_repeticao)
print("\nCategorias únicas:", categorias_unicas)
print("Total de categorias:", len(categorias_unicas))
````

## 7. Funções

Funções permitem **reutilizar código** sem repetição — um princípio fundamental de boas práticas.

```python
def nome_da_funcao(parametros):
    """Docstring: explica o que a função faz."""
    # corpo da função
    return resultado
```

````python
# Função simples com parâmetro e retorno
def classificar_nota(nota):
    """Classifica uma nota em Excelente, Aprovado, Recuperação ou Reprovado."""
    if nota >= 9:
        return "Excelente"
    elif nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

# Chamando a função para diferentes valores
for n in [9.5, 7.0, 5.5, 3.0]:
    print(f"Nota {n:4.1f} → {classificar_nota(n)}")
````

````python
# Parâmetros com valor padrão (default)
def calcular_media(notas, arredondar=2):
    """Calcula a média de uma lista de notas.

    Args:
        notas (list): lista de notas numéricas.
        arredondar (int): número de casas decimais (padrão: 2).

    Returns:
        float: média arredondada.
    """
    if len(notas) == 0:
        return 0.0
    return round(sum(notas) / len(notas), arredondar)

notas_turma = [8.5, 6.0, 9.0, 5.5, 7.5]

print("Média (2 casas):", calcular_media(notas_turma))
print("Média (0 casas):", calcular_media(notas_turma, arredondar=0))
````

````python
# Função que retorna múltiplos valores (retorna uma tupla internamente)
def resumo_estatistico(notas):
    """Retorna média, mínimo e máximo de uma lista de notas."""
    media  = calcular_media(notas)
    minimo = min(notas)
    maximo = max(notas)
    return media, minimo, maximo

notas_turma = [8.5, 6.0, 9.0, 5.5, 7.5]
media, minimo, maximo = resumo_estatistico(notas_turma)  # desempacota a tupla

print(f"Média: {media} | Mínimo: {minimo} | Máximo: {maximo}")
````

Código bem escrito é tão importante quanto código que funciona. Em Ciência de Dados você vai compartilhar notebooks com colegas e clientes.

O guia oficial de estilo do Python chama-se **PEP 8**. Os pontos principais:

## 9. Mini-Projeto: Análise de Desempenho de Alunos

Vamos integrar tudo o que aprendemos para analisar o desempenho de uma turma usando apenas Python puro — sem Pandas por enquanto.

**Objetivo:** calcular estatísticas, classificar alunos e exibir um relatório formatado.

````python
# Dataset: turma de alunos com nome, nota final e frequência
turma = [
    {"nome": "Ana",    "nota_final": 8.5, "frequencia": 90},
    {"nome": "Bruno",  "nota_final": 6.0, "frequencia": 75},
    {"nome": "Carla",  "nota_final": 9.0, "frequencia": 95},
    {"nome": "Diego",  "nota_final": 5.5, "frequencia": 68},
    {"nome": "Elaine", "nota_final": 7.5, "frequencia": 88},
    {"nome": "Felipe", "nota_final": 4.0, "frequencia": 60},
    {"nome": "Gia",    "nota_final": 9.5, "frequencia": 98},
    {"nome": "Hugo",   "nota_final": 3.5, "frequencia": 50},
]

print(f"Total de alunos: {len(turma)}")
````

````python
# Constantes — sem números mágicos no código
NOTA_MINIMA       = 7.0
FREQUENCIA_MINIMA = 75.0

def classificar_aluno(nota, frequencia):
    """Classifica um aluno considerando nota E frequência."""
    if frequencia < FREQUENCIA_MINIMA:
        return "Reprovado por falta"
    elif nota >= 9:
        return "Excelente"
    elif nota >= NOTA_MINIMA:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

def calcular_estatisticas(lista_notas):
    """Retorna média, mínimo e máximo."""
    media  = sum(lista_notas) / len(lista_notas)
    return round(media, 2), min(lista_notas), max(lista_notas)

print("Funções definidas com sucesso.")
````

````python
# Gera o relatório da turma
todas_notas = []

print("=" * 60)
print(f"{'ALUNO':<10} {'NOTA':>6} {'FREQ':>6}  SITUAÇÃO")
print("=" * 60)

for aluno in turma:
    situacao = classificar_aluno(aluno["nota_final"], aluno["frequencia"])
    todas_notas.append(aluno["nota_final"])
    print(f"{aluno['nome']:<10} {aluno['nota_final']:>6.1f} {aluno['frequencia']:>5}%  {situacao}")

print("=" * 60)
````

````python
# Estatísticas finais
media, minimo, maximo = calcular_estatisticas(todas_notas)

# Contagens por situação
aprovados   = [a for a in turma if classificar_aluno(a["nota_final"], a["frequencia"]) in ("Aprovado", "Excelente")]
recuperacao = [a for a in turma if classificar_aluno(a["nota_final"], a["frequencia"]) == "Recuperação"]
reprovados  = [a for a in turma if "Reprovado" in classificar_aluno(a["nota_final"], a["frequencia"])]

print("\n📊 RESUMO DA TURMA")
print("-" * 30)
print(f"Média da turma : {media}")
print(f"Nota mínima    : {minimo}")
print(f"Nota máxima    : {maximo}")
print("-" * 30)
print(f"Aprovados      : {len(aprovados)} aluno(s)")
print(f"Em recuperação : {len(recuperacao)} aluno(s)")
print(f"Reprovados     : {len(reprovados)} aluno(s)")
print("-" * 30)
taxa = len(aprovados) / len(turma) * 100
print(f"Taxa de aprovação: {taxa:.1f}%")
````
